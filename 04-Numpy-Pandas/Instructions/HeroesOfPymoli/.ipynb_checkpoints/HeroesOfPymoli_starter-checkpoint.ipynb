## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Total Count: male 484, female 81, other = 11: Percentage of Players: male 84.03, female 14.06, other 1.91

uniquesn_df = pymoli_df.drop_duplicates(subset='SN', keep='first', inplace=False)
uniquesn_df.head(10)

malecount = uniquesn_df[pymoli_df['Gender']=='Male'].count().mean()
femalecount = uniquesn_df[pymoli_df['Gender']=='Female'].count().mean()
totalcount = uniquesn_df.count().mean()
othercount = totalcount - malecount - femalecount
maleperc = malecount/totalcount
femaleperc = femalecount/totalcount
otherperc = othercount/totalcount


gendata = {'Category': ['Male', 'Female', 'Other'], 'Counts': [malecount, femalecount, totalcount], 'Percent': [maleperc, femaleperc, otherperc]}

gender_df = pd.DataFrame(data=gendata)
gender_df.head()

#femalecount
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchasecount: Female 113, male 652, other 15 | Avg Purchase Price 3.2, 3.02, 3.35 | Total Purchase Value 361, 1967, 50, avg total purchase per person 4.47, 4.07, 4.56
malepcount = pymoli_df[pymoli_df['Gender']=='Male'].count().mean()
femalepcount = pymoli_df[pymoli_df['Gender']=='Female'].count().mean()
totalpcount = pymoli_df['Gender'].count().mean()
otherpcount = pymoli_df[(pymoli_df['Gender']!='Female') & (pymoli_df['Gender']!='Male')].count().mean()

maleavgp = pymoli_df['Price'][pymoli_df['Gender']=='Male'].mean()
femaleavgp = pymoli_df['Price'][pymoli_df['Gender']=='Female'].mean()
otheravgp = pymoli_df['Price'][(pymoli_df['Gender']!='Female') & (pymoli_df['Gender']!='Male')].mean()

maleptotal = pymoli_df['Price'][pymoli_df['Gender']=='Male'].sum()
femaleptotal = pymoli_df['Price'][pymoli_df['Gender']=='Female'].sum()
otherptotal = pymoli_df['Price'][(pymoli_df['Gender']!='Female') & (pymoli_df['Gender']!='Male')].sum()

avgppersonmale = maleptotal/malecount
avgppersonfemale = femaleptotal/femalecount
avgppersonother = otherptotal/othercount

purchase_data = {'Gender': ['Male', 'Female', 'Other'], 'Purchases': [malepcount, femalepcount, otherpcount], 
                 'Avg Purchase Amount': [maleavgp, femaleavgp, otheravgp], 'Purchase Total': [maleptotal, femaleptotal, otherptotal],
                'Avg. Purchase Total per player': [avgppersonmale, avgppersonfemale, avgppersonother]}
purchase_df = pd.DataFrame(data=purchase_data)
purchase_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]

# Create the names for the four bins
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

uniquesn_df["Age band"] = pd.cut(uniquesn_df["Age"], bins, labels=bin_names)
age_df = uniquesn_df.groupby("Age band")
age_df2 = age_df.count()
age_df2.head()
totalplayers = age_df2["Age"].count().sum()
age_df2["Percentage Makeup"] = age_df["Age"].count()/totalplayers
age_df2.rename(columns={'Purchase ID': 'Counts'})
age_df3 = age_df2[['Purchase ID', 'Percentage Makeup']].head(10)
age_df3['Counts'] = age_df3['Purchase ID']
age_df3[['Percentage Makeup', 'Counts']].head(20)

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
players_df = pymoli_df.groupby("SN")
purchaseamounts = players_df["Price"].sum()
avgpurchaseamount = players_df["Price"].mean()
playerpurchases = players_df["Purchase ID"].count()

player_analysis = pd.DataFrame({"Purchase Count": playerpurchases, 
                             "Average Purchase Price": avgpurchaseamount, 
                             "Total Spend": purchaseamounts})
pa2 = player_analysis.sort_values(by=['Total Spend'],ascending=False)
pa2.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = pymoli_df.groupby("Item Name")
itemspend = items_df["Price"].sum()
avgitemprice = items_df["Price"].mean()
numofpurchases = items_df["Purchase ID"].count()

item_analysis = pd.DataFrame({"Purchase Count": numofpurchases, 
                             "Average Item Price": avgitemprice, 
                             "Total Spend": itemspend})

ia2 = item_analysis.sort_values(by=['Purchase Count'],ascending=False)
ia2.head(10)
```
