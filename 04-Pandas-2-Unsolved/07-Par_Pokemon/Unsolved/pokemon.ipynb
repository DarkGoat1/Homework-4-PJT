```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# File to load
file = 'Resources/Pokemon.csv'
```

```python
# Read with Pandas
df_poke = pd.read_csv(file, encoding="utf-8")
df_poke.head()
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
df_poke.loc[:, ["Name", "Type 1", "Type 2", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]].head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
df_poke["avg total"] = df_poke["Total"]/6
df_poke.head()
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.

df_type =  df_poke.groupby(['Type 1'])
print(df_type)
type_summary = df_type["Total", "avg total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"].mean()
type_sort = type_summary.sort_values(["Total"], ascending=False)
type_sort.head(18)
```
