```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Create reference to CSV file
csv_path = "Resources/Soccer2018Data.csv"

# Import the CSV into a pandas DataFrame
soccer_df = pd.read_csv(csv_path, low_memory=False)
soccer_df
```

```python
# Collect a list of all the unique values in "Preferred Position"
soccer_df["Preferred Position"].value_counts()
```

```python
# Looking only at strikers (ST) to start
only_strikers = soccer_df.loc[soccer_df["Preferred Position"] == "ST", :]
print(only_strikers)
```

```python
# Sort the DataFrame by the values in the "ST" column to find the worst
sorted_df = only_strikers.sort_values(['Overall'], ascending=[True])
sorted_df

# Reset the index so that the index is now based on the sorting locations
new_index = sorted_df.reset_index(drop=True)
new_index
```

```python
# Save all of the information collected on the worst striker
worst5_df = new_index.head(5)
worst5_df.to_csv('worst5players', sep='\t', encoding='utf-8')
```
